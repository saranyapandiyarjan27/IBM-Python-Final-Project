### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
# Identify missing values for all columns
missing_values = df.isnull().sum()
print("Missing values for all columns:")
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting missing data heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')  # cmap can be changed for different color schemes
plt.title('Missing Data Heatmap')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
# Count missing values in the 'Employment' column
missing_employment = df['Employment'].isnull().sum()

print(f"Number of missing values in 'Employment' column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `RemoteWork`).


```python
# Identify the most frequent value in the 'RemoteWork' column
most_frequent_value = df['RemoteWork'].mode()[0]  # [0] to get the first most frequent value
#also this code is efficient: freq_remote = df['RemoteWork'].value_counts().idxmax()
print(f"The most frequent value in 'RemoteWork' column is: {most_frequent_value}")
```

##### Task 6: Impute missing values in the `RemoteWork` column with the most frequent value.



```python
df['RemoteWork'].fillna(most_frequent_value, inplace=True)
# Verify the changes, previoulsly it has 10631 missing values.
print(df['RemoteWork'].isnull().sum())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `RemoteWork`).


```python
## Write your code here
plt.figure(figsize=(10,6))
df["RemoteWork"].value_counts().plot(kind='bar',color='skyblue', edgecolor='black')
plt.xlabel('Remote Work')
plt.xticks(rotation=0)
plt.ylabel('Count')
plt.title("Distribution of values in RemoteWork")
plt.show()
```
